##### Importaremos a biblioteca scipy utilizada para computação científica

```
#importar bibliotecas
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```
# Criar um numpy array entre -3 e +3
x = np.arange(-3,3,0.001)
```

```
# Plotar uma função densidade de probabilidade com distribuição normal.
plt.plot(x,norm.pdf(x), label='Distri A')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
```

```
# Incluir uma outra função com média 1.0 e desvio padrão de 0.5
plt.plot(x,norm.pdf(x), label='Distri A')
plt.plot(x,norm.pdf(x,1.0,0.5), label='Distri B')
plt.plot(x,0.5+norm.pdf(x,1.0,0.5), label='Distri B')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
```

```
# ajustes manuais nos eixos - limites do eixo x e y
axes = plt.axes()
axes.set_xlim([-4,4])
axes.set_ylim([-0.2,1.4])

plt.plot(x,norm.pdf(x), label='Distri A')
plt.plot(x,norm.pdf(x,1.0,0.5), label='Distri B')
plt.plot(x,0.5+norm.pdf(x,1.0,0.5), label='Distri B')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
```

```
# Vamos retirar algumas marcações no eixo X e Y e incluir a grade ticks


axes = plt.axes()
axes.set_xlim([-4,4])
axes.set_ylim([-0.2,1.4])
axes.set_xticks([-3,-1,1,2,3])
axes.set_yticks([0.4,0.8,1.3])


plt.plot(x,norm.pdf(x), label='Distri A')
plt.plot(x,norm.pdf(x,1.0,0.5), label='Distri B')
plt.plot(x,0.5+norm.pdf(x,1.0,0.5), label='Distri B')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.grid()
plt.legend()
# realizar o plot novamente
```

```
# variações nas cores e estilo das linhas
axes = plt.axes()
axes.set_xlim([-4,4])
axes.set_ylim([-0.2,1.4])
axes.set_xticks([-3,-1,1,2,3])
axes.set_yticks([0.4,0.8,1.3])


plt.plot(x,norm.pdf(x),'y--' )
plt.plot(x,norm.pdf(x,1.0,0.5),'b:')
plt.plot(x,0.5+norm.pdf(x,1.0,0.5),'g-.')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.grid()
plt.legend(['Distri A','Distri DF','Distri C'])
#outro modo de criar legendas
```
