## Ferramentas de Visualização - Scatter Plot da Iris (flor)


Vamos testar agora um outro exemplo de Scatter Plot usando um dos datasets mais populares para quem está estudando data science. Esse dataset foi publicado em 1936 por Ronald Fisher e foi muito utilizado em artigos, livros e pesquisas para demonstrar vários aspectos de data science e data mining.

O dataset possui 50 exemplos de cada uma das três espécies da flor de Íris, chamadas de setosa, virginica e versicolor. Cada exemplo tem quatro atributos diferentes: petal_length, petal_width, sepal_length e sepal_width. 

A base de dados IRIS pode ser encontrada em vários lugares e ser carregada de diversas maneiras.

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.dpi'] = 150
sns.set()
plt.rcParams['figure.figsize']=[5,5]
df = sns.load_dataset('iris')
```

```
sns.scatterplot(x=df["sepal_length"], y=df["sepal_width"], hue = df.species)
plt.xlabel('Septal Length')
plt.ylabel('Petal length')
plt.rcParams['figure.figsize']=[5,5]
```

```
iris = sns.load_dataset("iris")
ax = sns.scatterplot(x=iris.sepal_length, y=iris.petal_width,
                     hue=iris.species, style=iris.species)
```

```
# Utilizar subplot para mostrar 4 gráficos comparativos das features
plt.rcParams['figure.figsize']=[8,8]
plot= plt.subplot(2,2,1)
ax = sns.scatterplot(x=iris.sepal_length, y=iris.petal_width,
                     hue=iris.species, style=iris.species)
plot= plt.subplot(2,2,2)
ax = sns.scatterplot(x=iris.sepal_width, y=iris.petal_width,
                     hue=iris.species, style=iris.species)
plot= plt.subplot(2,2,3)
ax = sns.scatterplot(x=iris.sepal_width, y=iris.petal_length,
                     hue=iris.species, style=iris.species)
plot= plt.subplot(2,2,4)
ax = sns.scatterplot(x=iris.sepal_length, y=iris.petal_length,
                     hue=iris.species, style=iris.species)
```
