## Ferramentas de Visualização - Stacked Plot - Gráfico Empilhado

O gráfico empilhado possui esse nome pois ele marca de uma cor toda a área abaixo de uma linha e empilha diveros plots, um no topo do outro. Pode ser bastante útil quando nós precisamos visualizar o efeito acumulativo de múltiplas variáveis sendo plotadas no eixo Y.

````
# gastos com energia elétrica

# gastos com alimentação

# combustível e transporte

# cartao de credito
energia_eletrica = [220, 310, 280, 330, 300, 305, 295, 290, 340, 400, 370, 365]
alimentacao = [1450, 1400, 1380, 1500, 1600, 1800, 1350, 1399, 1400, 1500, 1450, 1990]
transporte = [782, 900, 732, 892, 334, 222, 300, 800, 900, 582, 596, 222]
cartao_credito = [700, 701, 702, 703, 704, 705, 706, 707, 708, 709, 710, 711]
````

Importe as bibliotecas necessárias para a parte gráfica

````
import matplotlib.pyplot as plt 
````

````
months= [x for x in range(1,13)]
# Create placeholders for plot and add required color 
plt.plot([],[], color='r', label='Energia Elétrica') 
plt.plot([],[], color='g', label='Alimentação') 
plt.plot([],[], color='b', label='Transporte') 
plt.plot([],[], color='y', label='Cartão de Crédito')
# Add stacks to the plot
plt.rcParams['figure.figsize']=[10,8]
plt.stackplot(months, energia_eletrica, alimentacao, transporte, cartao_credito, colors=['r', 'g', 'b', 'y'])
plt.legend()
# Add Labels 
plt.title('Gastos da Casa') 
plt.xlabel('Meses do Ano')
plt.ylabel('Custo')
````

Outro Exemplo utilizando dados na estrutura de dictionary<br>



````
import numpy as np
import matplotlib.pyplot as plt

# data from United Nations World Population Prospects (Revision 2019)
# https://population.un.org/wpp/, license: CC BY 3.0 IGO
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2018]
population_by_continent = {
    'africa': [228, 284, 365, 477, 631, 814, 1044, 1275],
    'americas': [340, 425, 519, 619, 727, 840, 943, 1006],
    'asia': [1394, 1686, 2120, 2625, 3202, 3714, 4169, 4560],
    'europe': [220, 253, 276, 295, 310, 303, 294, 293],
    'oceania': [12, 15, 19, 22, 26, 31, 36, 39],
}

fig, ax = plt.subplots()
ax.stackplot(year, population_by_continent.values(),
             labels=population_by_continent.keys())
ax.legend(loc='upper left')
ax.set_title('World population')
ax.set_xlabel('Year')
ax.set_ylabel('Number of people (millions)')
````
