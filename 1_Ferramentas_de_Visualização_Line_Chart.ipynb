# LINE CHART

### O line chart é um tipo de gráfico que mostra as informações como uma série de pontos conectados por segmentos de linhas. 
#### É um tipo básico  comum em muitas áreas. É geralmente utilizado para visualizar tendências em um intervalo de tempo, ou seja, uma série temporal.

```
# bibliotecas necessárias
import matplotlib.pyplot as plt
import numpy as np
# criar dados artificaias através do random
values = np.cumsum(np.random.randn(10,1))
# realizar o plot básico
plt.rcParams['figure.figsize']=(8,8)
plt.plot(values)
```

```
values2 = np.cumsum(np.random.randn(10,1))
```

```
# modificar linha com linestyle '-', dashed , '--', '-.', ':'
plt.plot(values, linestyle=':')
```

```
# modificar linha com linewidth
plt.plot(values, linestyle=':', linewidth=3 )
```

```
# modificar linha com color
plt.plot(values, linestyle=':', linewidth=5, color='black' )
plt.plot(values2, linestyle='-', linewidth=3, color='g' )
```

```
#modificar linha com o marker

plt.plot(values, linestyle=':', linewidth=1, color='black', marker='o')
plt.plot(values2, linestyle='-', linewidth=1, color='g', marker='*')
```

```
# modificar eixos e títulos
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('Line Chart Example')
plt.plot(values, linestyle=':', linewidth=1, color='black', marker='o')
plt.plot(values2, linestyle='-', linewidth=1, color='g', marker='*')
```

```
# legendas
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('Line Chart Example')
plt.plot(values, label='primeira serie', linestyle=':', linewidth=1, color='black', marker='o')
plt.plot(values2, label='segunda serie', linestyle='-', linewidth=1, color='g', marker='*')
plt.legend()
```
