## Ferramentas de Visualização - Gráfico de Pizza - Pie Chart

Esse tipo de gráfico causa bastante polêmica entre os especialistas e pesquisadores. <br>
Em 2005, Ian Spence publicou um artigo argumentando que o gráfico de pizza (ou torta, dependendo da preferência) não deveria ser utilizado. **O motivo é que o ser humano tem dificuldade em distinguir diferenças entre pedaços de um círculo.** Além disso, as pessoas tem a tendencia de superestimar o tamanho dos angulos mais abertos e subsestimar os angulos mais fechados( agudos) de cada pedaço. <br>
**Há um contra-argumento**, apesar do gráfico não ser tão comunicativo como deveria, **as pessoas adoram círculos**. Por isso acabam sempre pensando nesse tipo de gráfico como solução.

### Usaremos o dataset dos Pokemos para desenhar o nosso pie chart.
Esse dataset está disponível no Kaggle em: https://www.kaggle.com/rounakbanik/pokemon <br>
Fizemos uma cópia no nosso github também: 
https://raw.githubusercontent.com/aprendadatascience/visualization-tools/master/datasets/pokemon.csv

```
url = 'https://raw.githubusercontent.com/aprendadatascience/visualization-tools/master/datasets/pokemon.csv'
```

```
import pandas as pd
df = pd.read_csv(url)
```

Importe as bibliotecas necessárias para a parte gráfica

```
df = df[['name','type1']]
```

```
df = pd.DataFrame(df['type1'].value_counts())
```

```
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[10,10]
plt.pie(df['type1'], labels=df.index, shadow=False, startangle=90, autopct='%1.1f%%',)
plt.axis('equal')
plt.show()
```
