# LINE CHART

### O line chart é um tipo de gráfico que mostra as informações como uma série de pontos conectados por segmentos de linhas. 
#### É um tipo básico  comum em muitas áreas. É geralmente utilizado para visualizar tendências em um intervalo de tempo, ou seja, uma série temporal.

```
# bibliotecas necessárias
import matplotlib.pyplot as plt
import numpy as np
 
# criar dados artificaias através do random
values=np.cumsum(np.random.randn(1000,1))

# realizar o plot básico
plt.plot(values)
```

```
# modificar linha com linestyle '-', dashed , '--', '-.', ':'
values=np.cumsum(np.random.randn(100,1))
plt.plot(values, linestyle='-.')
```

```
import pandas as pd
df=pd.DataFrame({'xvalues': range(91,101), 'yvalues': np.random.randn(10) })
 
# plot
plt.plot( 'xvalues', 'yvalues', data=df)
```

```
# modificar linha com linewidth
plt.plot(values, linewidth=5)
```

```
# modificar linha com color
plt.plot(values, linewidth=2.5, color='r')
```

```
#modificar linha com o marker
#plt.plot(values, linewidth=2.5, color='r', marker='o')
#plt.plot(values, linewidth=2.5, color='r', marker='v')
plt.rcParams['figure.figsize'] = (14, 10)
plt.plot(values, linewidth=2.5, color='r', marker='*')
```

```
# modificar eixos e títulos
plt.xlabel('X numbers')
plt.ylabel('Y numbers')
plt.title('Line Chart')
plt.plot(values, linewidth=2.5, color='r', marker='*')
```

```
# legendas
x1 = [1, 2, 3]
y1 = [4, 7, 5]

x2 = [1, 2, 3]
y2 = [10, 14, 12]

plt.plot(x1, y1, label='First line', color='c')  # c - cyan
plt.plot(x2, y2, label='Second line', color='red')

plt.xlabel('X numbers')
plt.ylabel('Y numbers')
plt.title('Line Chart')
plt.legend()

plt.show()
```

```
import datetime
import math
import pandas as pd
import random
import radar
from faker import Faker
   
fake = Faker()

def generateData(n):
     listdata = []
     start = datetime.datetime(2019, 3, 1)
     end = datetime.datetime(2020, 7, 30)
     delta = end - start
     for _ in range(n):
       date = radar.random_datetime(start='2019-03-1',
                                    stop='2020-07-30').strftime("%Y-%m-%d")
       price = round(random.uniform(950, 1000), 4)
       listdata.append([date, price])
     df = pd.DataFrame(listdata, columns = ['Date', 'Price'])
     df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
     df = df.groupby(by='Date').mean()
     return df
```
